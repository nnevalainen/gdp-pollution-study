```python
import os
from IPython.display import display

import pandas as pd
import numpy as np
```

```python
DATA_DIR = os.path.join('..', 'data', 'gdp_data')

countires = [os.path.join(DATA_DIR, country) for country in os.listdir(DATA_DIR)]
csvs = [pd.read_csv(country, sep=',', skiprows=3) for country in countires]

DATA = pd.concat(csvs, axis=0)

data = DATA[DATA['Indicator Name'] == 'GDP per capita (current US$)']
data = data.set_index('Country Code', drop=True)
data = data.drop(['Unnamed: 62', 'Indicator Name', 'Indicator Code'], axis=1)
```

```python
with open('head.tex', 'w') as f:
    f.write(data.iloc[:10, pd.np.r_[0:4, 55:59]].to_latex())
```
